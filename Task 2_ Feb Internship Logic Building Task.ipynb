# 1: Unique Words in a Sentence


Problem Statement

Given a sentence, identify all unique words and display:

The count of unique words

The unique words themselves


```python
# Taking input sentence from user
sentence = input("Enter sentence: ")

# Splitting sentence into words
words = sentence.split()

# Converting list to set to remove duplicates
unique_words = set(words)

# Printing results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

# 2: Highest Salary from Employee Data



Problem Statement

Store employee names and their salaries, and determine which employee has the highest salary.


```python
# Storing employee names and their salaries in a dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Assuming first salary as highest (initially 0)
highest_salary = 0
highest_name = ""

# Looping through dictionary to check each salary
for name in employees:
    
    # Get salary of current employee
    salary = employees[name]
    
    # Compare with highest salary
    if salary > highest_salary:
        highest_salary = salary
        highest_name = name

# Printing the final result
print("Highest Salary:", highest_name, "-", highest_salary)
```

# 3: Find Maximum and Minimum Values


Problem Statement

Find the maximum and minimum values from a list of numbers.


```python
# List of numbers
numbers = [45, 22, 89, 10, 66]

# Assume first element as max and min initially
max_value = numbers[0]
min_value = numbers[0]

# Loop through each number in the list
for num in numbers:
    
    # Check for maximum value
    if num > max_value:
        max_value = num
    
    # Check for minimum value
    if num < min_value:
        min_value = num

# Printing results
print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

# 4: Count Products Above a Price Threshold


Problem Statement

Count how many products have a price greater than 1000.


```python
# List of product prices
prices = [450, 1200, 899, 1500, 300]

# Variable to count products above 1000
count = 0

# Loop through each price
for price in prices:
    
    # Check if price is greater than 1000
    if price > 1000:
        count += 1

# Print final result
print("Products above 1000:", count)
```

# 5: Calculate Attendance Percentage


Problem Statement

Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).



```python
# List showing attendance (P = Present, A = Absent)
attendance = ["P", "P", "A", "P", "P"]

# Variable to count present days
present_count = 0

# Loop through attendance list
for day in attendance:
    
    # Check if the student was present
    if day == "P":
        present_count += 1

# Total number of days
total_days = len(attendance)

# Calculate attendance percentage
percentage = (present_count / total_days) * 100

# Print result
print("Attendance Percentage:", percentage)
```

# 7: Count Character Frequency


Problem Statement

Count how many times each character appears in a given string.


```python
# Given string
text = "python"

# Empty dictionary to store character frequency
char_count = {}

# Loop through each character in the string
for ch in text:
    
    # If character already in dictionary, increase count
    if ch in char_count:
        char_count[ch] += 1
    else:
        # If character not in dictionary, add it with value 1
        char_count[ch] = 1

# Print final result
print(char_count)
```

# 10: Calculate Average Marks


Problem Statement

Calculate the average marks of a student from a list of marks.


```python
# List of marks
marks = [70, 75, 80, 65]

# Variable to store total
total = 0

# Loop through each mark to calculate sum
for mark in marks:
    total += mark

# Calculate average
average = total / len(marks)

# Print result
print("Average Marks:", average)
```
