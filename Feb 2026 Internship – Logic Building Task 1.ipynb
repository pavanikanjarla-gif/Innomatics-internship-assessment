2. Pass / Fail Analyzer
   
Problem Statement:
Analyze student results.
marks = [45, 78, 90, 33, 60]


```python
# List of marks obtained by students
marks = [45, 78, 90, 33, 60]

# Variables to count pass and fail students
pass_count = 0
fail_count = 0

# Checking each mark one by one
for m in marks:
    # If marks are 50 or above, student passes
    if m >= 50:
        pass_count += 1
    else:
        # Otherwise student fails
        fail_count += 1

# Printing the final result
print("Total Students:", len(marks))
print("Students Passed:", pass_count)
print("Students Failed:", fail_count)
```

3. Simple Data Cleaner
   
Problem Statement:
Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]


```python
# Original list of names
names = [" Alice ", "bob", " CHARLIE "]

# Empty list to store cleaned names
cleaned_names = []

# Cleaning each name
for name in names:
    # Remove spaces and convert to lowercase
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

# Print final cleaned list
print("Cleaned Names:", cleaned_names)
```

4. Message Length Analyzer
   
Problem Statement:
Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]


```python
# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Checking each message
for msg in messages:
    length = len(msg)
    
    print("Message:", msg)
    print("Length:", length)
    
    # Flag if message is longer than 10 characters
    if length > 10:
        print("Status: Long Message")
    
    print()  # Blank line for clarity
```

5. Error Message Detector
   
Problem Statement:
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]


```python
# List of system log messages
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Variable to count error messages
error_count = 0

# Checking each log entry
for log in logs:
    if log == "ERROR":
        error_count += 1

# Printing total error count
print("Total ERROR messages:", error_count)
```
